```python
%matplotlib inline
import pickle
import numpy as np
import matplotlib.pyplot as plt

from model import lorenz63_fdm
from model import M63 as M
from assimilation import M4DVar
```

```python
X_nature = np.load('./data/X_nature.npy')
X_ini = np.load('./data/X_ini.npy')
ts = np.load('./data/time_span.npy')
Pb = np.load('./data/Pb.npy')
R = np.load('./data/R.npy')
```

```python
dt = 0.01
```

```python
def RMSE(forecast, nature):
    return np.sqrt(np.mean((forecast-nature)**2, axis=0))
```

# Assimilate normal observation

#### `rmse_fdv` as CNTL in the later experiments

```python
win_len = 6
fcs_len = 10
win_num = ts.size // (win_len+fcs_len)

X_obs = np.load('./data/4DVar/obs_normal.npy')
X_obs.shape
```

```python
fdv = M4DVar(lorenz63_fdm, dt)
fdv.list_params()
params = {
    'X_ini': X_ini,
    'obs': X_obs,
    'window_num': win_num,
    'window_len': win_len,
    'forecast_len': fcs_len,
    'Pb': Pb,
    'R': R,
    'H_func': lambda arr: arr,
    'H': np.eye(3),
    'M': lambda X: M(X.ravel(), dt),
}
fdv.set_params(**params)
fdv.cycle()
```

```python
rmse_fdv = RMSE(fdv.analysis, X_nature)

plt.figure(figsize=(10, 5))
plt.plot(ts, rmse_fdv)
plt.xlabel('Time')
plt.ylabel('RMSE')
#plt.ylim([-0.2, 10])
plt.title('4DVar analysis RMSE')
print(rmse_fdv.mean())
```

# First moment

### Assimilate bias observations based on normality test

```python
# first moment
ex_obs_dict = pickle.load(open('./data/4DVar/obs_bias_005_040.pickle', 'rb'))
ex_obs_dict.keys()
```

```python
plt.figure(figsize=(12, 6))

mean_rmse = []        # mean RMSE
mean_rmse_nospn = []  # mean RMSE without spin up periods
var_rmse = []
var_rmse_nospn = []

for obsm in ex_obs_dict.keys():
    obs = ex_obs_dict[obsm]

    fdv = M4DVar(lorenz63_fdm, dt)
    fdv.list_params()
    params = {
        'X_ini': X_ini,
        'obs': obs,
        'window_num': win_num,
        'window_len': win_len,
        'forecast_len': fcs_len,
        'Pb': Pb,
        'R': R,
        'H_func': lambda arr: arr,
        'H': np.eye(3),
        'M': lambda X: M(X.ravel(), dt),
    }
    fdv.set_params(**params)
    fdv.cycle()

    # calculate RMSE and stable index
    rmse = RMSE(fdv.analysis, X_nature)
    print(
        f'obsm={obsm}, Mean RMSE={rmse.mean():.4f}, Variance RMSE={rmse.var():.4f},  '
        + f'Without spinup: Mean RMSE={rmse[50:].mean():.4f}, Variance RMSE={rmse[50:].var():.4f}'
    )
    mean_rmse.append(rmse.mean())
    mean_rmse_nospn.append(rmse[50:].mean())
    var_rmse.append(rmse.var())
    var_rmse_nospn.append(rmse[50:].var())

    plt.plot(ts, rmse, label=obsm)
    
plt.plot(ts, rmse_fdv, 'k', label='CNTL')
plt.legend()
plt.title('RMSE')
plt.xlabel('Time')
plt.ylabel('rmse')
plt.ylim([-0.2, 10]);
```

### Repeated assimilate skew observation based on normality test (repeat 10 times)

```python
ex_obs_dict = pickle.load(open('./data/4DVar/obs_skew_015_135.pickle', 'rb'))
ex_alpha = ex_obs_dict.keys()
ex_alpha = list(map(float, ex_alpha))
ex_alpha
```

```python
from generate_obs_skew import gen_skewnormal

def gen_skew_obs_4dvar(X_nature, mean, var, alpha, size, win_len, fcs_len, win_num):
    X_obs_err = np.zeros((3, ts.size))
    for irow in range(3):
        # generate observations
        skew_obs = gen_skewnormal(mean, var, alpha, size=size).reshape((-1, win_len))
        skew_obs = np.hstack([skew_obs, np.zeros((fcs_len*win_num)).reshape((-1, fcs_len))])
        skew_obs = skew_obs.ravel()
        X_obserr[irow,:] = skew_obs
    idx = np.all(X_obserr != 0, axis=0)
    X_nature_tmp = X_nature.copy()
    X_nature_tmp[:,idx] = 0
    ex_obs = X_nature - X_nature_tmp + X_obserr
    return ex_obs
```

```python
rmse_dict = {}
for ex_a in ex_alpha:
    print(ex_a)
    rmse = np.zeros_like(ts)
    
    for _ in range(10):
        obs = gen_skew_obs_4dvar(X_nature, 0, 2, ex_a, win_len*win_num, win_len, fcs_len, win_num)
        
        fdv = M4DVar(lorenz63_fdm, dt)
        fdv.list_params()
        params = {
            'X_ini': X_ini,
            'obs': obs,
            'window_num': win_num,
            'window_len': win_len,
            'forecast_len': fcs_len,
            'Pb': Pb,
            'R': R,
            'H_func': lambda arr: arr,
            'H': np.eye(3),
            'M': lambda X: M(X.ravel(), dt),
        }
        fdv.set_params(**params)
        fdv.cycle()
        rmse += RMSE(fdv.analysis, X_nature)

    rmse /= 10
    key = f'{ex_a:.2f}'
    rmse_dict[key] = rmse
```

```python
plt.figure(figsize=(12, 6))
plt.plot(ts, rmse_fdv, 'k', label='CNTL')

for key, rmse in rmse_dict.items():
    print(f'alpha = {key}, mean RMSE = {rmse.mean():.4f}')
    plt.plot(ts, rmse, label=key)

plt.legend()
plt.title('Mean RMSE')
plt.xlabel('Time')
plt.ylabel('RMSE');
```

### Check the ratio of skew observation located between -2*std ~ 2*std

```python
ex_obs_dict = pickle.load(open('./data/4DVar/obs_skew_130_330.pickle', 'rb'))
ex_obs_dict.keys()
```

```python
nrows = len(ex_obs_dict.keys())
fig, axs = plt.subplots(nrows=nrows, figsize=(5, 20), sharex=True)

for i, (key, obs) in enumerate(ex_obs_dict.items()):
    idx = np.all(obs != 0, axis=0)
    obserr = (obs[:,idx] - X_nature[:,idx]).ravel()
    axs[i].hist(obserr)
    
    std = np.sqrt(2)
    axs[i].vlines([2*std, -2*std], ymin=0, ymax=200, ls='dashed')
    axs[i].set_xlim([-4, 6])
    axs[i].set_title(f'alpha = {key}')
    
    total_size = obserr.size
    within_size = obserr[np.where((obserr <= 2*std) & (obserr >= -2*std))].size
    ratio = within_size / total_size
    axs[i].text(-1, 175, f'ratio={ratio:.2f}', fontsize=13)

plt.tight_layout()
```
